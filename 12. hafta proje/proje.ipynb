table

```python
import seaborn as sns

class KuyumcuUygulamasi:
    def init(self):
        self.veri = None

    def veri_al(self):
        try:
            self.veri = sns.load_dataset("diamonds")
            print("Veri başarıyla yüklendi!")
        except Exception as e:
            print("Hata: Veri yüklenirken bir sorun oluştu.")
            print(e)

    def fiyat_hesapla(self):
        if self.veri is not None:
            carat = float(input("Carat değerini girin: "))
            cut = input("Kesim türünü girin (Fair, Good, Very Good, Premium, Ideal): ")
            color = input("Renk sınıfını girin (D, E, F, G, H, I, J): ")
            clarity = input("Berraklık sınıfını girin (I1, SI2, SI1, VS2, VS1, VVS2, VVS1, IF): ")

            filtrelenmis_veri = self.veri[(self.veri["carat"] == carat) &
                                          (self.veri["cut"] == cut) &
                                          (self.veri["color"] == color) &
                                          (self.veri["clarity"] == clarity)]

            if len(filtrelenmis_veri) > 0:
                fiyat = filtrelenmis_veri["price"].mean()
                print(f"Elmasın tahmini fiyatı: ${fiyat:.2f}")
            else:
                print("Girdiğiniz kriterlere uygun elmas bulunamadı.")

    def proje_baslat(self):
        self.veri_al()
        if self.veri is not None:
            self.fiyat_hesapla()

# Ana program
uygulama = KuyumcuUygulamasi()
uygulama.proje_baslat()

```

```python
import seaborn as sns
import numpy as np


class DiamondStats:
   def __init__(self):
       self.diamonds = sns.load_dataset('diamonds')
       self.cut_order = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
       self.color_order = ['J', 'I', 'H', 'G', 'F', 'E', 'D']
       self.clarity_order = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']


   def ozetDataset(self):
       ozet = self.diamonds.describe().T
       return ozet


   def kesimeGoreOrtUcret(self):
       kesime_gore_ort_ucret = self.diamonds.groupby('cut').mean()['price']
       return kesime_gore_ort_ucret


   def rengeGoreOrtUcret(self):
       renge_gore_ort_ucret = self.diamonds.groupby('color').mean()['price']
       return renge_gore_ort_ucret


   def berraklıgaGoreOrtUcret(self):
       berraklıga_gore_ort_ucret = self.diamonds.groupby('clarity').mean()['price']
       return berraklıga_gore_ort_ucret


def main():
   ds = DiamondStats()
   print("Pırlanta Dünyasına Hoşgeldiniz!")
   while True:
       print("Lütfen yapmak için bir işlem seçin:")
       print("1. Pırlanta istatistikleri için özet tablosu")
       print("2. Kesime  göre ortalama ücret")
       print("3. Renge göre ortalama ücret")
       print("4. Berraklık derecesine göre ortalama ücret")
       print("5. Çıkış")
       try:
           option = int(input("Option: "))
           if option == 1:
               summary = ds.ozetDataset()
               print(summary)
           elif option == 2:
               mean_price_by_cut = ds.kesimeGoreOrtUcret()
               print(mean_price_by_cut)
           elif option == 3:
               mean_price_by_color = ds.rengeGoreOrtUcret()
               print(mean_price_by_color)
           elif option == 4:
               mean_price_by_clarity = ds.berraklıgaGoreOrtUcret()
               print(mean_price_by_clarity)
           elif option == 5:
               print("Exiting program...")
               break
           else:
               print("Geçerli olmayan işlem seçtiniz!")
       except ValueError:
           print("Geçerli olmayan giriş yaptınız! Lütfen bir sayı giriniz.")
           continue


if __name__ == '__main__':
   main()

```

```python
class AltinTahmini:
    def __init__(self):
        self.veri = None
    
    def veri_al(self):
        altin = yf.Ticker("XAUUSD=X")
        self.veri = altin.history(period="1d")

    def fiyat_hesapla(self):
        if self.veri is not None:
            kalite = input("Altın kalitesini girin (24 ayar, 22 ayar, 18 ayar): ")
            agirlik = float(input("Altın ağırlığını girin (gram): "))
            iscilik = input("İşçilik türünü girin (sade, yüzük, bilezik): ")

            if kalite == "24 ayar":
                fiyat_gr = 250
            elif kalite == "22 ayar":
                fiyat_gr = 220
            elif kalite == "18 ayar":
                fiyat_gr = 180
            else:
                print("Geçersiz kalite seçimi.")
                return

            if iscilik == "sade":
                fiyat_iscilik = 0
            elif iscilik == "yüzük":
                fiyat_iscilik = 50
            elif iscilik == "bilezik":
                fiyat_iscilik = 100
            else:
                print("Geçersiz işçilik seçimi.")
                return

            fiyat = (agirlik * fiyat_gr) + fiyat_iscilik
            print(f"Altının tahmini fiyatı: ${fiyat:.2f}")
        else:
            print("Hata: Veri yüklenirken bir sorun oluştu.")

    def baslat(self):
        self.veri_al()
        self.fiyat_hesapla()

# Ana program
tahminci = AltinTahmini()
tahminci.baslat()
```

```python
import requests

def altin_fiyati():
    try:
        url = "https://altin.doviz.com/gram-altin"
        response = requests.get(url)
        response.raise_for_status()
        data = response.json()
        return data["buying"]
    except:
        print("Altın fiyatı alınamadı.")
        return None

def altin_fiyat_hesapla(kalite, agirlik, iscilik_maliyeti):
    gram_altin_fiyati = altin_fiyati()
    if gram_altin_fiyati is None:
        return None
    if kalite == "külçe":
        fiyat = agirlik * gram_altin_fiyati
    elif kalite in ["küpe", "yüzük", "bilezik", "çeyrek"]:
        fiyat = (agirlik * gram_altin_fiyati) + iscilik_maliyeti
    elif kalite == "gr":
        fiyat = agirlik * gram_altin_fiyati
    elif kalite == "cumhuriyet":
        fiyat = (agirlik * gram_altin_fiyati) + (iscilik_maliyeti * 0.5)
    else:
        print("Geçersiz altın kalitesi.")
        return None
    return fiyat

kalite = input("Altın kalitesini girin (külçe, küpe, yüzük, bilezik, çeyrek, gr, cumhuriyet): ")
agirlik = float(input("Altın ağırlığını girin (gram cinsinden): "))

if kalite in ["küpe", "yüzük", "bilezik", "çeyrek", "cumhuriyet"]:
    iscilik_maliyeti = float(input("İşçilik maliyetini girin: "))

fiyat = altin_fiyat_hesapla(kalite, agirlik, iscilik_maliyeti)

if fiyat is not None:
    print(f"Altının tahmini fiyatı: {fiyat:.2f} TL")
```
