```python
import requests, time, socket, sys
from bs4 import BeautifulSoup
import datetime

# Şu anki tarih ve saat bilgisi
bugun = datetime.datetime.now()
# Tarih formatını ayarlayalım
tarih_formati = "{}.{}.{}".format(bugun.day, bugun.month, bugun.year)
# Saat formatını ayarlayalım
saat_formati = "{}:{}".format(bugun.hour, bugun.minute)
# Gram Altının fiyatını İnternet Sitesinden Çekiyoruz.
url = "https://bigpara.hurriyet.com.tr/altin/gram-altin-fiyati/"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
gram_altin = soup.find("span", {"class": "value up"})
gram_altin = gram_altin.text.strip().replace('.','').replace(',','.')
gram_altin=float(gram_altin)
# Çeyrek Altının fiyatını İnternet Sitesinden Çekiyoruz.
url = "https://bigpara.hurriyet.com.tr/altin/ceyrek-altin-fiyati/"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
ceyrek_altin = soup.find("span", {"class": "value up"})
ceyrek_altin = ceyrek_altin.text.strip().replace('.','').replace(',','.')
ceyrek_altin=float(ceyrek_altin)
# Yarım Altının fiyatını İnternet Sitesinden Çekiyoruz.
url = "https://bigpara.hurriyet.com.tr/altin/yarim-altin-fiyati/"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
yarim_altin = soup.find("span", {"class": "value up"})
yarim_altin = yarim_altin.text.strip().replace('.','').replace(',','.')
yarim_altin=float(yarim_altin)
# Tam Altının fiyatını İnternet Sitesinden Çekiyoruz.
url = "https://bigpara.hurriyet.com.tr/altin/tam-altin-fiyati/"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
tam_altin = soup.find("span", {"class": "value up"})
tam_altin = tam_altin.text.strip().replace('.','').replace(',','.')
tam_altin=float(tam_altin)
# Cumhuriyet Altının fiyatını İnternet Sitesinden Çekiyoruz.
url = "https://bigpara.hurriyet.com.tr/altin/cumhuriyet-altini-fiyati/"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
cumhuriyet_altin = soup.find("span", {"class": "value up"})
cumhuriyet_altin = cumhuriyet_altin.text.strip().replace('.','').replace(',','.')
cumhuriyet_altin=float(cumhuriyet_altin)
# 22 Ayar Altının fiyatını İnternet Sitesinden Çekiyoruz.
url = "https://bigpara.hurriyet.com.tr/altin/22-ayar-altin-fiyati/"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
a_22_altin = soup.find("span", {"class": "value up"})
a_22_altin = a_22_altin.text.strip().replace('.','').replace(',','.')
a_22_altin=float(a_22_altin)
# 18 Ayar Altının fiyatını İnternet Sitesinden Çekiyoruz.
url = "https://bigpara.hurriyet.com.tr/altin/18-ayar-altin-fiyati/"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
a_18_altin = soup.find("span", {"class": "value up"})
a_18_altin = a_18_altin.text.strip().replace('.','').replace(',','.')
a_18_altin=float(a_18_altin)
# 14 Ayar Altının fiyatını İnternet Sitesinden Çekiyoruz.
url = "https://bigpara.hurriyet.com.tr/altin/14-ayar-altin-fiyati/"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
a_14_altin = soup.find("span", {"class": "value up"})
a_14_altin = a_14_altin.text.strip().replace('.','').replace(',','.')
a_14_altin=float(a_14_altin)
  
while True:
   print("*****************************")
   print("YetGen Kuyumculuk Sistemi'ne Hoş Geldiniz")
   print("*****************************\n")
   print("Lütfen yapmak için bir işlem seçin:")
   print("1. Güncel Altın Fiyatları\n2. Hazır Altın Ör: Gram,Yarım,Çeyrek...\n3. Ziynet Eşyası Ör: Kolye,Bilezik,Küpe\n4. Hurda-Hasarlı Altın\n5. Çıkış")
   try:
       option = int(input("Seçiminiz: "))
       if option == 1: #güncel altın fiyatları
           print("\n")
           print("{} tarihi, saat {} itibariyla güncel altın fiyatları:".format(tarih_formati, saat_formati))
           print(f"Gram altın: {gram_altin}TL")
           print(f"Çeyrek altın: {ceyrek_altin}TL")
           print(f"Yarım altın: {yarim_altin}TL")
           print(f"Tam altın: {tam_altin}TL")
           print(f"Cumhuriyet altını: {cumhuriyet_altin}TL")
           print(f"22 ayar altın: {a_22_altin}TL")
           print(f"18 ayar altın: {a_18_altin}TL")
           print(f"14 ayar altın: {a_14_altin}TL")
          
       elif option == 2: ##Hazır altınlar: adet seçme işlemi ile gram, çeyrek, yarım, tam, cumhuriyet altını seçimi yapılacak
           print("Elinde Hangi Altından Var?")
           print("1. Gram Altın\n2. Çeyrek Altın\n3. Yarım Altın\n4. Tam Altın\n5. Cumhuriyet Altını\n6. Geri Dön")
           try:
               hazir_altin_secim = int(input("Seçiminiz: "))
               if hazir_altin_secim == 1:
                   adet = int(input("Lütfen kaç adet gram altın satmak istediğinizi girin: "))
                   toplam_fiyat = adet * gram_altin
                   print(f"{adet} gram altın için sana biçtiğim toplam fiyat: {toplam_fiyat}TL")                 
               elif hazir_altin_secim == 2:
                   adet = int(input("Lütfen kaç adet çeyrek altın satmak istediğinizi girin: "))
                   toplam_fiyat1 = adet * ceyrek_altin
                   print(f"{adet} çeyrek altın için sana biçtiğim toplam fiyat:: {toplam_fiyat1}TL")                 
               elif hazir_altin_secim == 3:
                   adet = int(input("Lütfen kaç adet yarım altın satmak istediğinizi girin: "))
                   toplam_fiyat2 = adet * yarim_altin
                   print(f"{adet} yarım altın için sana biçtiğim toplam fiyat:: {toplam_fiyat2}TL")                 
               elif hazir_altin_secim == 4:
                   adet = int(input("Lütfen kaç adet tam altın satmak istediğinizi girin: "))
                   toplam_fiyat3 = adet * tam_altin
                   print(f"{adet} tam altın için sana biçtiğim toplam fiyat:: {toplam_fiyat3}TL")           
               elif hazir_altin_secim == 5:
                   adet = int(input("Lütfen kaç adet cumhuriyet altını satmak istediğinizi girin: "))
                   toplam_fiyat4 = adet * cumhuriyet_altin
                   print(f"{adet} cumhuriyet altını için sana biçtiğim toplam fiyat: {toplam_fiyat4}TL")
               elif hazir_altin_secim == 6:
                   print("...Menüye Dönülüyor...")
                   continue
           except ValueError:
               print("Geçerli olmayan giriş yaptınız! Lütfen bir sayı giriniz.")
            
       elif option == 3: #ziynet eşyası: kolye, bilezik, küpe, yüzük ten biri seçilecek; altının ayarı: 14, 18, 22 den biri seçilecek; gram değeri girilecek; işçilik: kötü {%12}, orta {%22}, iyi {%36}, çok iyi {%36}, mükemmel {%100} den biri seçilecek
           # örneğin: kolye, 14 ayar, 20 gr, iyi işçilik seçilsin = ({a_14_altin} * girilen gram değeri) * %136 = fiyat biçildi
           altin_turleri = ["14 ayar", "18 ayar", "22 ayar"]
           iscilik_turleri = ["kötü", "orta", "iyi", "çok iyi", "mükemmel"]
           print("Aşağıdaki altın türlerinden birini seçiniz:")
           for index, altin in enumerate(altin_turleri, start=1):
               print(f"{index}. {altin}")
           altin_secim = int(input("Seçiminizi yapınız: "))
           while altin_secim not in range(1, len(altin_turleri) + 1):
               altin_secim = int(input("Geçersiz seçim. Lütfen tekrar deneyiniz: "))
           secilen_altin = altin_turleri[altin_secim - 1]
           gram = float(input("Eşyanın gram değerini giriniz: "))
           print("Aşağıdaki işçilik türlerinden birini seçiniz:")
           for index, iscilik in enumerate(iscilik_turleri, start=1):
               print(f"{index}. {iscilik}")
           iscilik_secim = int(input("Seçiminizi yapınız: "))
           while iscilik_secim not in range(1, len(iscilik_turleri) + 1):
               iscilik_secim = int(input("Geçersiz seçim. Lütfen tekrar deneyiniz: "))
           secilen_iscilik = iscilik_turleri[iscilik_secim - 1]
           altin_fiyat = {
               "14 ayar": a_14_altin,
               "18 ayar": a_18_altin,
               "22 ayar": a_22_altin
           }
           iscilik_oran = {
               "kötü": 0.2,
               "orta": 0.5,
               "iyi": 1,
               "çok iyi": 1.9,
               "mükemmel": 2.5
           }
           fiyat = (float(altin_fiyat[secilen_altin]) * gram) * (1 + iscilik_oran[secilen_iscilik])
           print(f"Ürünün fiyatı: {fiyat:.2f} TL")
              
       elif option == 4:  # eyşyanın gramı ölçülecek, altının ayarı : 14, 18, 22, 24 den biri seçilip gram fiyatına göre %20 ucuza satılacak
           print("Lütfen satılacak ziynet eşyasının ayrısını giriniz:")
           print("1- 14 Ayar")
           print("2- 18 Ayar")
           print("3- 22 Ayar")
           print("4- 24 Ayar")
           altin_ayari_secim = int(input("Seçiminiz: "))
           gram = float(input("Lütfen satılacak ziynet eşyasının gram miktarını giriniz: "))
           if altin_ayari_secim == 1:
               fiyat = a_14_altin * gram * 0.8
               print(f"Toplam satış fiyatı: {fiyat} TL")
           elif altin_ayari_secim == 2:
               fiyat = a_18_altin * gram * 0.8
               print(f"Toplam satış fiyatı: {fiyat} TL")
           elif altin_ayari_secim == 3:
               fiyat = a_22_altin * gram * 0.8
               print(f"Toplam satış fiyatı: {fiyat} TL")
           elif altin_ayari_secim == 4:
               fiyat = gram_altin * gram * 0.8
               print(f"Toplam satış fiyatı: {fiyat} TL")
           else:
               print("Geçersiz bir seçim yaptınız!")

       elif option == 5:
           print("...Program Kapanıyor...")
           time.sleep(2)
           print("Program kapandı.")
           break
       else:
           print("Geçerli olmayan işlem seçtiniz!")
   except ValueError:
       print("Geçerli olmayan giriş yaptınız! Lütfen bir sayı giriniz.")
       continue
```
