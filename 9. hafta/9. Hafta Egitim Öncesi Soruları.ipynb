# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
class Vehicle:
    def __init__(self, marka, motor_hacmi):
        self.marka = marka
        self.motor_hacmi= motor_hacmi
        print("Vehicle nesnesi türetildi")
    
    def intro(self):
        print(self.marka, self.motor_hacmi)

class Bus(Vehicle):
    def __init__(self, marka, motor_hacmi, kapi):
        super().__init__(marka, motor_hacmi)
        self.kapi = kapi
        print("Bus nesnesi çalıştı")
    
    def intro(self):
        print(self.marka, self.motor_hacmi, self.kapi)
        
    def kapasite(self, koltuk=50):
        print(f"The seating capacity of a bus is {koltuk} passengers")


v1 = Vehicle("Mercedes", 120)
v1.intro()

print("\n")

b1 = Bus("Volvo", 90, 2)
b1.intro()
b1.kapasite()


````
