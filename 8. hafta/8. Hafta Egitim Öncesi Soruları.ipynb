# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
file = open("urunler.txt", "w", encoding="utf-8")

ürünAdi = input("Ürünün adını giriniz")
ürünFiyati = input("Ürünün fiyatını giriniz")

file.write(ürünAdi)
file.write(" \n")
file.write(ürünFiyati)
file.close()
````

ve değiştirilmesini istediğimiz kelimemizi yerini alıcak olan yeni kelime isimli değişkene atayın. Belirlenen kelimeleri text dosyasında değiştiren fonksiyonu yazınız.

````python
eskiKelime = input("Bir kelime giriniz")
yeniKelime = "hahahhah"

with open("urunler.txt", "r+", encoding="utf-8") as file:
    liste = file.readlines()

    liste.insert(3, eskiKelime)
    file.write(" \n")
    file.write(yeniKelime)
    file.seek(0)
    file.writelines(liste)
````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
class vehicle():
    def __init__(self, Name, Speed, Mileage):
        self.Name = Name
        self.Speed = Speed
        self.Mileage = Mileage

Bus = vehicle("School Volvo", 180, 12)
comments = [Bus]

for i in comments:
    print(f"{i.Name} marka araç, maksimum {i.Speed} km/h hıza sahiptir. Aynı zamanda  {i.Mileage} mil yol gidebilir.")
````
